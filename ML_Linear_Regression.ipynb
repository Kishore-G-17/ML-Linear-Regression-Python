```python
#Modules to import for perform Linear regression on the given dataset.
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Read the '.csv' file 
df = pd.read_csv("area.csv")
df
```

```python
#Scatter the data points in the graph to get the basic idea of what model we need to use.
%matplotlib inline
plt.xlabel('Area(sqr ft)')
plt.ylabel('Prices(INR)')
plt.scatter(df.Area, df.Price, color="red", marker="+")
```

```python
#Create and train the linear regression model
lm = linear_model.LinearRegression()
lm.fit(df[['Area']], df.Price)
```

```python
#plot our linear regression model classifier in the graph with the given data points.
%matplotlib inline
plt.xlabel('Area(sqr ft)')
plt.ylabel('Prices(INR)')
plt.scatter(df.Area, df.Price, color="red", marker="+")
plt.plot(df.Area, lm.predict(df[['Area']]), color="blue")
```

```python
#Now we can ust our linear regression model to predict the output
lm.predict([[3000]])
```

```python
lm.coef_
```

```python
lm.intercept_
```
